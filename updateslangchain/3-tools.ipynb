```python
##tools

"""models can requests to call tools that perform tasks such as fetching data from a database, 
    searching the web or running code. Tools are pairing of:
    1. A schema, including the name of the tools, a descroption, 
    and/or argument definition(often json schema)
    2. A function or continue to excecute"""
```

```python
import os
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model

load_dotenv()

print(os.getenv("###"))   # should show key, not None

model = init_chat_model(
    "llama-3.3-70b-versatile",
    model_provider="groq"
)

response = model.invoke("What is the capital of US?")

print(response.content)
```

```python
from langchain.tools import tool
from langchain.chat_models import init_chat_model

@tool
def get_weather(location:str) -> str:
    """get the weather at a location"""
    return f"It sunny in {location}"

model_with_tools = model.bind_tools([get_weather])
response = model_with_tools.invoke("what is the weather in India?")
print(response)

for tool_call in response.tool_calls:
    print(f"Tool: {tool_call['name']}")
    print(f"Args: {tool_call['args']}")
```
